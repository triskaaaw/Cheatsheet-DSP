```python
# Set transactions
gp_invoiceno = dataset.groupby('InvoiceNo')
transactions = []
for name,group in gp_invoiceno:
    transactions.append(list(group['Description'].map(str)))
```

# Algoritma Apriori

```python
import time

# Measure time for Apriori algorithm
start_time_apriori = time.time()

# Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, min_support = 0.01, min_confidence = 0.5, min_lift = 3, min_length = 2)
# Get the results
results = list(rules)

end_time_apriori = time.time()
apriori_time = end_time_apriori - start_time_apriori
```

```python
# Visualising the result as a readeable dataframe
final_results_apriori = pd.DataFrame(np.random.randint(low=0, high=1, size=(len(results), 6)), columns=['GeneralRules', 'LeftRules', 'RightRules', 'Support', 'Confidence', 'Lift'])
index = 0
for g, s, i in results:
    final_results_apriori.iloc[index] = [' _&&_ '.join(list(g)), ' _&&_ '.join(list(i[0][0])), ' _&&_ '.join(list(i[0][1])), s, i[0][2], i[0][3]]
    index = index+1
# The most significant rules
final_results_apriori = final_results_apriori.sort_values('Lift', ascending=0)
final_results_apriori.head(20)
```

```python
count=1
for i, d in final_results_apriori.iterrows():
    print('Rule #'+str(count)+':')
    print(d['LeftRules'])
    print('=> '+d['RightRules'])
    print('Support: '+str(d['Support'])+' - Confidence: '+str(d['Confidence'])+' - Lift: '+str(d['Lift']))
    print('--------------------')
    count=count+1
```

# Algoritma FP Growth

```python
# Import necessary libraries
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Measure time for FP-Growth algorithm
start_time_fpgrowth = time.time()

# Create a DataFrame with one-hot encoding
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply FP-Growth algorithm
frequent_itemsets = fpgrowth(df, min_support=0.01, use_colnames=True)

# Generate the association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Filter rules based on the lift and minimum length of antecedents
filtered_rules = rules[(rules['lift'] >= 3) & (rules['antecedents'].apply(lambda x: len(x)) >= 2)]

end_time_fpgrowth = time.time()
fpgrowth_time = end_time_fpgrowth - start_time_fpgrowth
```

```python
# Prepare the final results DataFrame
final_results_fpgrowth = pd.DataFrame(columns=['Antecedents', 'Consequents', 'Support', 'Confidence', 'Lift'])

# Fill the DataFrame with filtered rules
for index, row in filtered_rules.iterrows():
    antecedents = ' _&&_ '.join(list(row['antecedents']))
    consequents = ' _&&_ '.join(list(row['consequents']))
    support = row['support']
    confidence = row['confidence']
    lift = row['lift']
    new_row = pd.DataFrame({
        'Antecedents': [antecedents],
        'Consequents': [consequents],
        'Support': [support],
        'Confidence': [confidence],
        'Lift': [lift]
    })
    final_results_fpgrowth = pd.concat([final_results_fpgrowth, new_row], ignore_index=True)

# Sort the DataFrame by Support
final_results_fpgrowth = final_results_fpgrowth.sort_values(by='Support', ascending=False)

# Display the top 400 results
print(final_results_fpgrowth.head(400))
```

```python
count=1
for i, d in final_results_fpgrowth.iterrows():
    print('Rule #'+str(count)+':')
    print(d['Antecedents'])
    print('=> '+d['Consequents'])
    print('Support: '+str(d['Support'])+' - Confidence: '+str(d['Confidence'])+' - Lift: '+str(d['Lift']))
    print('--------------------')
    count=count+1
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from apyori import apriori
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
from memory_profiler import memory_usage


# Function to run Apriori algorithm
def run_apriori():
    apriori_rules = apriori(transactions, min_support=0.01, min_confidence=0.5, min_lift=3, min_length=2)
    apriori_results = list(apriori_rules)
    return apriori_results

# Function to run FP-Growth algorithm
def run_fpgrowth():
    te = TransactionEncoder()
    te_ary = te.fit(transactions).transform(transactions)
    df = pd.DataFrame(te_ary, columns=te.columns_)
    frequent_itemsets = fpgrowth(df, min_support=0.01, use_colnames=True)
    fpgrowth_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
    filtered_rules = fpgrowth_rules[(fpgrowth_rules['lift'] >= 3) & (fpgrowth_rules['antecedents'].apply(lambda x: len(x)) >= 2)]
    return filtered_rules

# Measure memory usage for Apriori
apriori_memory = memory_usage((run_apriori,), interval=0.1)
apriori_max_memory = max(apriori_memory)

# Measure memory usage for FP-Growth
fpgrowth_memory = memory_usage((run_fpgrowth,), interval=0.1)
fpgrowth_max_memory = max(fpgrowth_memory)

# Display the memory usage
print(f"Apriori algorithm max memory usage: {apriori_max_memory:.2f} MiB")
print(f"FP-Growth algorithm max memory usage: {fpgrowth_max_memory:.2f} MiB")
```

```python
import matplotlib.pyplot as plt

# Nama algoritma dan penggunaan memori maksimum
algorithms = ['Apriori', 'FP-Growth']
memory_usage = [apriori_max_memory, fpgrowth_max_memory]

# Membuat bar plot
plt.figure(figsize=(10, 6))
plt.bar(algorithms, memory_usage, color=['blue', 'green'])

# Menambahkan judul dan label
plt.title('Comparison of Max Memory Usage for Apriori and FP-Growth Algorithms')
plt.xlabel('Algorithm')
plt.ylabel('Memory Usage (MiB)')

# Menambahkan teks pada setiap bar
for i, memory in enumerate(memory_usage):
    plt.text(i, memory + 0.5, f'{memory:.2f}', ha='center', va='bottom')

# Menampilkan plot
plt.show()
```
